**Pengelompokan Data**




```
print(df["Jenis Iris Flower"].unique())
```

```
kolom = ['Lebar kelopak daun', 'Panjang kelopak daun', 'Lebar kelopak', 'Panjang kelopak']
print(kolom)
```

```
grup=df[['Jenis Iris Flower'] + kolom].groupby(['Jenis Iris Flower'])
grup.head(2)
```

**Membuat Tabel Pivot**

```
pivot = df.pivot_table(index='Jenis Iris Flower', values=kolom, aggfunc='mean')
print(pivot)
```

**Hitung korelasi**

```
for i in range(len(kolom)):
    for j in range(i + 1, len(kolom)):
        pearson_coef, p_value = stats.pearsonr(df[kolom[i]], df[kolom[j]])

        print(f"Korelasi antara {kolom[i]} dan {kolom[j]}:")
        print(f"Koefisien Korelasi Pearson : {pearson_coef}")
        print(f"P-value: {p_value}\n")
```

**Heatmap**

**Anova**

```
for atribut in kolom:
  groups = df['Jenis Iris Flower'].unique()
  data = [df[atribut][df['Jenis Iris Flower'] == g] for g in groups]
  fvalue, pvalue = stats.f_oneway(*data)
  print(f"ANOVA {atribut}:")
  print(f"F-value: {fvalue}")
  print(f"P-value: {pvalue}\n")
```
